```python
origins = dist.loc[:,["origin", "origin_location"]
            ].groupby(["origin", "origin_location"]).count().reset_index()
origins.columns = ["stop", "stop_location"]
destinations = dist.loc[:,["destination", "destination_location"]
            ].groupby(["destination", "destination_location"]).count().reset_index()
destinations.columns = ["stop", "stop_location"]
stops = origins.append(destinations, ignore_index = True)
# extract location and normalize to integers
stops["longitude"] = (pd.to_numeric(stops.stop_location.apply(lambda x: x.strip().split(",")[0]
                                                            )) * 1000000).astype(int)
stops["longitude"] -= stops["longitude"].min()
stops["latitude"] = (pd.to_numeric(stops.stop_location.apply(lambda x: x.strip().split(",")[1]
                                                           )) * 1000000).astype(int)
stops["latitude"] -= stops["latitude"].min()
stops.drop_duplicates(inplace = True)
```

```python
# new dist_matrix
def add_dist(r):
  return r["distance"] + int(r["new_num"].split("-")[-1])
def add_dur(r):
  return r["duration"] + int(r["new_num"].split("-")[-1])

# originates from indices
new_org = dist.loc[dist.origin.isin(indices),:].merge(
    n_stops.loc[:,["stop", "new_num"]], left_on = "origin", right_on = "stop")
new_org.distance = new_org.apply(add_dist, axis = 1)
new_org.duration = new_org.apply(add_dur, axis = 1)
new_org.origin = new_org.new_num
# destination are indices
new_dest = dist.loc[dist.destination.isin(indices),:].merge(
    n_stops.loc[:,["stop", "new_num"]], left_on = "destination", right_on = "stop")
new_dest.distance = new_dest.apply(add_dist, axis = 1)
new_dest.duration = new_dest.apply(add_dur, axis = 1)
new_dest.destination = new_dest.new_num
# remove from dist, and append new rows
n_dist = dist.loc[~dist.destination.isin(indices) & ~dist.origin.isin(indices),:].copy()
n_dist = pd.concat([n_dist, new_org])
n_dist = pd.concat([n_dist, new_dest])
```

```python
def get_total_distance_duration(res, dist_mat, time_mat, demands):
    total_distance = 0
    total_duration = 0  # Duration in minutes
    total_demand = 0    # Total demand

    routes = res.best.get_routes()
    for route in routes:
        spots = route.visits()
        route_distance = 0
        route_duration = 0  # Duration in minutes
        route_demand = 0    # Demand for this route

        # Calculate distance, duration, and demand from depot to the first spot
        route_distance += dist_mat[0, spots[0]]
        route_duration += time_mat[0, spots[0]] + 10
        route_demand += demands[spots[0]]

        # Calculate distance, duration, and demand for each leg of the route, adding 10 minutes per leg
        for i in range(len(spots) - 1):
            route_distance += dist_mat[spots[i], spots[i + 1]]
            route_duration += time_mat[spots[i], spots[i + 1]] + 10
            route_demand += demands[spots[i + 1]]

        # Add distance and duration from last spot back to depot
        route_distance += dist_mat[spots[-1], 0]
        route_duration += time_mat[spots[-1], 0] + 10

        total_distance += route_distance
        total_duration += route_duration
        total_demand += route_demand

        print(f"Route {route}: Distance = {route_distance / 1000} km, Duration = {route_duration / 60} hours, Demand = {route_demand}")

    print(f"Total Distance: {total_distance / 1000} km, Total Duration: {total_duration / 60} hours, Total Demand: {total_demand}")
    print(f"Average Distance: {total_distance / 1000 / len(routes)} km, Average Duration: {total_duration / 60 / len(routes)} hours, Average Demand: {total_demand / len(routes)}")
    return total_distance, total_duration, total_demand
```

```python
dist_mat = np.array(dist_mat)  # Assuming dist_mat is your distance matrix
dur_matt = np.array(dur_matt)  # Assuming time_mat is your time matrix

# Call the function to get total distance and duration
get_total_distance_duration(res, dist_mat, dur_matt)
```

```python
from networkx import DiGraph, relabel_nodes, set_node_attributes
import networkx as nx
from numpy import array
from vrpy import VehicleRoutingProblem as VRP

# Distance matrix
DISTANCES = new_dist.values.tolist()
TRAVEL_TIMES = dur_matt.values.tolist()

for i in range(num_nodes):
  DISTANCES[i][0] = 0
  TRAVEL_TIMES[i][0] = 0
  DISTANCES[num_nodes-1][i] = 0
  TRAVEL_TIMES[num_nodes-1][i] = 0
DISTANCES[0][num_nodes-1] = 0
TRAVEL_TIMES[0][num_nodes-1] = 0


# Time windows (key: node, value: lower/upper bound)
TIME_WINDOWS_LOWER = {i: 0 for i in range(len(DISTANCES))}
TIME_WINDOWS_LOWER[0] = 0
TIME_WINDOWS_UPPER = {i: 60*400 for i in range(1, len(DISTANCES))}
# TIME_WINDOWS_LOWER = {0: 0, 1: 7, 2: 10, 3: 16, 4: 10, 5: 0, 6: 5, 7: 0, 8: 5, 9: 0, 10: 10, 11: 10, 12: 0, 13: 5, 14: 7, 15: 10, 16: 11,}
# TIME_WINDOWS_UPPER = {1: 12, 2: 15, 3: 18, 4: 13, 5: 5, 6: 10, 7: 4, 8: 10, 9: 3, 10: 16, 11: 15, 12: 5, 13: 10, 14: 8, 15: 15, 16: 15,}

DEMAND = {i: demands[i] for i in range(1, len(DISTANCES))}

# Transform distance matrix into DiGraph
A = array(DISTANCES, dtype=[("cost", int)])
G_d = nx.from_numpy_array(A, create_using=DiGraph())

# Transform time matrix into DiGraph
A = array(TRAVEL_TIMES, dtype=[("time", int)])
G_t = nx.from_numpy_array(A, create_using=DiGraph())

# Merge
G = nx.compose(G_d, G_t)

# Set time windows
set_node_attributes(G, values=TIME_WINDOWS_LOWER, name="lower")
set_node_attributes(G, values=TIME_WINDOWS_UPPER, name="upper")

G = relabel_nodes(G, {0: "Source", num_nodes-1: "Sink"})

# The VRP is defined and solved
prob = VRP(G, load_capacity = 1200, num_vehicles = 30, drop_penalty = 10000)
prob.solve()
```

```python
def create_data_model():
    """Stores the data for the problem."""
    data = {}
    data["distance_matrix"] = dist
    data['time_matrix'] = dur


    data['demands'] = dmds

    # data["demands"] = [0, 1, 1, 3, 6, 3, 6, 8, 8, 1, 2, 1, 2, 6, 6, 8, 8]
    data["vehicle_capacities"] = [int(i[0]) for i in veh]
    data["num_vehicles"] = len(veh)
    data["depot"] = 0
    data['time_window_lower'] = [0, 0, 2, 4, 5, 7, 10, 12, 15, 18, 20, 22, 24, 25, 26, 28, 30]
    data['time_window_upper'] = [100] * 17
    data["time_windows"] = wds
    data['time_limit_seconds'] = 10
    return data
```

```python
def print_solution(data, manager, routing, solution, locations0):
    """Prints solution on console."""
    locations = locations0['origin_location']
    total_distance = 0
    total_load = 0
    time_dimension = routing.GetDimensionOrDie('Time')
    total_time = 0

    routes_res = []

    for vehicle_id in range(data['num_vehicles']):
        index = routing.Start(vehicle_id)
        plan_output = 'Route for vehicle {}:\n'.format(vehicle_id)
        route_load = 0
        rr = []
        while not routing.IsEnd(index):
            node_index = manager.IndexToNode(index)
            time_var = time_dimension.CumulVar(index)
            route_load += data['demands'][node_index]
            plan_output += 'Place {0:>2} Arrive at {2:>2}min Depart at {3:>2}min (Load {1:>2})\n'.format(manager.IndexToNode(index), route_load, solution.Min(time_var), solution.Max(time_var))
            rr.append(manager.IndexToNode(index))#####
            previous_index = index
            index = solution.Value(routing.NextVar(index))

        routes_res.append(rr)  ####

        time_var = time_dimension.CumulVar(index)
        total_time += solution.Min(time_var)
        plan_output +="Place {0:>2} Arrive at {2:>2}min \n\n".format(manager.IndexToNode(index), route_load, solution.Min(time_var), solution.Max(time_var))

        # route output
        plan_output += 'Load of the route: {}\n'.format(route_load)
        plan_output += 'Time of the route: {}min\n'.format(solution.Min(time_var))
        plan_output += "--------------------"

        print(plan_output)
        total_load += route_load

    print('Total load of all routes: {}'.format(total_load))
    print('Total time of all routes: {}min'.format(total_time))
    dropped_nodes = "Dropped nodes:"
    for node in range(routing.Size()):
        if routing.IsStart(node) or routing.IsEnd(node):
            continue
        if solution.Value(routing.NextVar(node)) == node:
            dropped_nodes += f" {manager.IndexToNode(node)}"
    print(dropped_nodes)

    for rt in routes_res:
      # if len(rt) > 1:
      #   route_coordinates = []
      #   for node in rt:
      #     locc = locations[node].split(',')
      #     route_coordinates.append((float(locc[0]), float(locc[1])))
      #   # route_coordinates = [(locations[node]['latitude'], locations[node]['longitude']) for node in rt]
      #   m = folium.Map(location=route_coordinates[0], zoom_start=13)

      #   # Add a marker for each coordinate
      #   for coord in route_coordinates:
      #       folium.Marker(location=coord).add_to(m)

      #   # Add lines to the map for each route segment
      #   folium.PolyLine(route_coordinates, color='blue').add_to(m)

      print(rt, '*****************')

    return routes_res


def main():
    """Solve the VRP with time windows."""
    # Instantiate the data problem.
    data = create_data_model()

    # Create the routing index manager.
    manager = pywrapcp.RoutingIndexManager(len(data['time_matrix']),
                                           data['num_vehicles'], data['depot'])

    # Create Routing Model.
    routing = pywrapcp.RoutingModel(manager)


    # Create and register a transit callback.
    def time_callback(from_index, to_index):
        """Returns the travel time between the two nodes."""
        # Convert from routing variable Index to time matrix NodeIndex.
        from_node = manager.IndexToNode(from_index)
        to_node = manager.IndexToNode(to_index)
        return data['time_matrix'][from_node][to_node]

    transit_callback_index = routing.RegisterTransitCallback(time_callback)

    # Define cost of each arc.
    routing.SetArcCostEvaluatorOfAllVehicles(transit_callback_index)

    # Add Capacity constraint.
    def demand_callback(from_index):
        """Returns the demand of the node."""
        # Convert from routing variable Index to demands NodeIndex.
        from_node = manager.IndexToNode(from_index)
        return data['demands'][from_node]

    demand_callback_index = routing.RegisterUnaryTransitCallback(
        demand_callback)
    routing.AddDimensionWithVehicleCapacity(
        demand_callback_index,
        0,  # null capacity slack
        data['vehicle_capacities'],  # vehicle maximum capacities
        True,  # start cumul to zero
        'Capacity')

    # Allow to drop nodes.
    # penalty = 1000
    # penalty = [100000000] * (len(data['demands'])) #+ [100000, 5000]
    # for node in range(1, len(data["distance_matrix"])):
    #     routing.AddDisjunction([manager.NodeToIndex(node)], penalty[node-1])


    # Add Time Windows constraint.
    time = 'Time'
    routing.AddDimension(
        transit_callback_index,
        30,  # allow waiting time
        10000,  # maximum time per vehicle # PROBLEM
        False,  # Don't force start cumul to zero.
        time)
    time_dimension = routing.GetDimensionOrDie(time)

    # Add time window constraints for each location except depot.
    for location_idx, time_window in enumerate(data['time_windows']):
        if location_idx == 0:
            continue
        index = manager.NodeToIndex(location_idx)
        time_dimension.CumulVar(index).SetRange(time_window[0], time_window[1])

    # Add time window constraints for each vehicle start node.
    for vehicle_id in range(data['num_vehicles']):
        index = routing.Start(vehicle_id)
        time_dimension.CumulVar(index).SetRange(data['time_windows'][0][0], data['time_windows'][0][1])

    # Instantiate route start and end times to produce feasible times.
    for i in range(data['num_vehicles']):
        routing.AddVariableMinimizedByFinalizer(
            time_dimension.CumulVar(routing.Start(i)))
        routing.AddVariableMinimizedByFinalizer(
            time_dimension.CumulVar(routing.End(i)))

    # Setting first solution heuristic.
    search_parameters = pywrapcp.DefaultRoutingSearchParameters()
    search_parameters.first_solution_strategy = (
        routing_enums_pb2.FirstSolutionStrategy.PATH_CHEAPEST_ARC)
    search_parameters.time_limit.seconds = data['time_limit_seconds']

    # Solve the problem.
    solution = routing.SolveWithParameters(search_parameters)

    # Print solution on console.
    if solution:
        routes_res = print_solution(data, manager, routing, solution, locations0)
    else:
      print('infeasible')
      return 'infeasible', False
    return solution, routes_res
```
