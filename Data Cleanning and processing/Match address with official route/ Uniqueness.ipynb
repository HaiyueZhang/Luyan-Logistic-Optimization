```python
df2 = pd.read_csv("Xiamen_locations最新有仓库.csv")
df2
```

```python
df2_unique = df2.drop_duplicates(subset='location', keep='first')
df2_unique
df2_unique.to_csv('Unique_locations最新有仓库.csv', index=False)
```

```python
index_map = df2.drop_duplicates(subset='location', keep='first').set_index('location')['Index'].to_dict()

# Apply the mapping to set 'Index' values
df2['Index'] = df2['location'].map(index_map)

# Identify and set the 'location' of duplicates to NaN, excluding the first occurrence
mask = df2.duplicated(subset='location', keep='first')
df2.loc[mask, 'location'] = np.nan
df2.to_csv('Xiamen_Unique_locations最新有仓库.csv')
```

```python
df3 = pd.read_csv("Xiamen_Unique_locations最新有仓库.csv")
```

```python
df3 = df3.sort_values(by="Index", ascending=True)
unique_vals = df3['Index'].unique()
mapping = {val: i for i, val in enumerate(unique_vals)}
# Use the mapping to transform the 'Index' column
df3['New_Index'] = df3['Index'].map(mapping)
df3 = df3.drop('Index', axis=1)
df3 = df3.rename(columns={'New_Index': 'Index'})
df3.to_csv('Xiamen_Unique_locations最新有仓库.csv')
```
