```python
# Kmeans + TSP(Greedy Algorithm) for each clusters
# Load the new address DataFrame
address_df = pd.read_csv("2023-07-18_Xiamen.csv")

# Extract unique locations and convert them to [Latitude, Longitude]
locations = address_df['location'].unique().tolist()
numeric_locations = [list(map(float, loc.split(',')))[::-1] for loc in locations]  # Reverse the order to [Latitude, Longitude]

# K-means Clustering
kmeans = KMeans(n_clusters=12, random_state=0).fit(numeric_locations)

clusters = {}
for i, label in enumerate(kmeans.labels_):
    if label not in clusters:
        clusters[label] = []
    clusters[label].append(numeric_locations[i])

tsp_routes = {}
for label, points in clusters.items():
    route = [points[0]]
    points = points[1:]
    while points:
        next_point = min(points, key=lambda point: np.linalg.norm(np.array(route[-1]) - np.array(point)))
        route.append(next_point)
        points.remove(next_point)
    tsp_routes[label] = route

# Initialize a map centered around the first location
m = folium.Map(location=numeric_locations[0], zoom_start=12)

# Define 12 distinct colors for the routes
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22', '#17becf','#AD00FF', "#FF9900"]
for label, route in tsp_routes.items():
    route = np.array(route)
    color = colors[label % len(colors)]
    
    # Add a polyline for the route and markers for each location
    folium.PolyLine(route, color=color, weight=5, opacity=1).add_to(m)
    for idx, point in enumerate(route):
        folium.CircleMarker(
            location=point, 
            radius=5,  # Adjust the size by setting the radius
            popup=f'Route {label + 1}, Order {idx + 1}', 
            color=color,
            fill=True,
            fill_color=color
        ).add_to(m)

# Display the map
m
```

```python
df_c12 = pd.read_csv("Clusters/0718_distance_cluster12.csv")
df_distance12 = pd.read_csv("Real Distance/trip_details_0718_distance_cluster12_20231009_1140.csv")
df_c12['Index'] = range(len(df_c12))

# Ensure the columns you are merging on are of the same data type
df_distance12['origin'] = df_distance12['origin'].astype(int)
df_distance12['destination'] = df_distance12['destination'].astype(int)
df_c12['Index'] = df_c12['Index'].astype(int)

# Now, perform the merge for 'origin'
df_distance12 = pd.merge(df_distance12, df_c12[['Index', 'location', '线路名称']], left_on='origin', right_on='Index', how='left')
df_distance12.rename(columns={'location': 'origin_location', '线路名称': 'origin_线路名称'}, inplace=True)
df_distance12.drop(columns='Index', inplace=True)

# Perform the merge for 'destination'
df_distance12 = pd.merge(df_distance12, df_c12[['Index', 'location', '线路名称']], left_on='destination', right_on='Index', how='left')
df_distance12.rename(columns={'location': 'destination_location', '线路名称': 'destination_线路名称'}, inplace=True)
df_distance12.drop(columns='Index', inplace=True)
df_distance12
```

```python
# Christofides algorithm(distance)
# Create a graph from the dataframe
G = nx.Graph()
for index, row in df_distance12.iterrows():
    G.add_edge(row['origin_location'], row['destination_location'], weight=row['distance'])
tsp_path = nx.approximation.traveling_salesman_problem(G, cycle=False)
start_location = df_distance12[df_distance12['origin'] == 0]['origin_location'].iloc[0]
if start_location in tsp_path:
    tsp_path = tsp_path[tsp_path.index(start_location):] + tsp_path[:tsp_path.index(start_location)]
location_tuples = [tuple(map(float, loc.split(','))) for loc in tsp_path]

# Visualization
m = folium.Map(location=location_tuples[0][::-1], zoom_start=13)
folium.PolyLine(locations=[loc[::-1] for loc in location_tuples], color="blue", weight=2.5, opacity=1).add_to(m)
for i, loc in enumerate(location_tuples):
    folium.Marker(location=loc[::-1], popup=f"Stop {i + 1}").add_to(m)

# Calculate total distance
total_distance = sum(G[tsp_path[i]][tsp_path[i + 1]]['weight'] for i in range(len(tsp_path) - 1))
print("Total distance of the TSP path:", total_distance)
m
# 51.853 km
```

```python
# Define the main Map
m = folium.Map(zoom_start=13)

# Define colors for each cluster path
colors = [
    'red', 'blue', 'green', 'purple', 'orange', 'darkred',
    'lightred', 'beige', 'darkblue', 'darkgreen', 'cadetblue', 'darkpurple'
]

# Loop through all clusters
for cluster_num in range(1, 13):  # Assuming there are 12 clusters

    # Read cluster and distance data
    df_c = pd.read_csv(f"Clusters/0718_distance_cluster{cluster_num}.csv")
    df_distance = pd.read_csv(f"Real Distance/trip_details_0718_distance_cluster{cluster_num}.csv")

    df_c['Index'] = range(len(df_c))

    # Ensure the columns you are merging on are of the same data type
    df_distance['origin'] = df_distance['origin'].astype(int)
    df_distance['destination'] = df_distance['destination'].astype(int)
    df_c['Index'] = df_c['Index'].astype(int)

    # Merge for 'origin' and 'destination'
    df_distance = pd.merge(df_distance, df_c[['Index', 'location', '线路名称']], left_on='origin', right_on='Index', how='left')
    df_distance.rename(columns={'location': 'origin_location', '线路名称': 'origin_线路名称'}, inplace=True)
    df_distance.drop(columns='Index', inplace=True)

    df_distance = pd.merge(df_distance, df_c[['Index', 'location', '线路名称']], left_on='destination', right_on='Index', how='left')
    df_distance.rename(columns={'location': 'destination_location', '线路名称': 'destination_线路名称'}, inplace=True)
    df_distance.drop(columns='Index', inplace=True)

    # Check if there's data in the dataframe
    if df_distance.empty:
        print(f"No data for cluster {cluster_num}. Skipping...")
        continue

    # Create a graph from the dataframe
    G = nx.Graph()
    for index, row in df_distance.iterrows():
        G.add_edge(row['origin_location'], row['destination_location'], weight=row['distance'])

    tsp_path = nx.approximation.traveling_salesman_problem(G, cycle=False)

    # Check for the start_location in the current cluster data
    if len(df_distance[df_distance['origin'] == 0]) > 0:
        start_location = df_distance[df_distance['origin'] == 0]['origin_location'].iloc[0]
        if start_location in tsp_path:
            tsp_path = tsp_path[tsp_path.index(start_location):] + tsp_path[:tsp_path.index(start_location)]
    else:
        print(f"No start location found for cluster {cluster_num}. Skipping this cluster.")
        continue

    location_tuples = [tuple(map(float, loc.split(','))) for loc in tsp_path]

    # Visualization for the current cluster
    folium.PolyLine(locations=[loc[::-1] for loc in location_tuples], color=colors[cluster_num - 1], weight=2.5, opacity=1).add_to(m)
    for i, loc in enumerate(location_tuples):
        folium.Marker(location=loc[::-1], popup=f"Cluster {cluster_num} Stop {i + 1}").add_to(m)

    # Calculate total distance for the current cluster
    total_distance = sum(G[tsp_path[i]][tsp_path[i + 1]]['weight'] for i in range(len(tsp_path) - 1))
    print(f"Total distance of the TSP path for cluster {cluster_num}: {total_distance}")

# Display the final map
m
```
