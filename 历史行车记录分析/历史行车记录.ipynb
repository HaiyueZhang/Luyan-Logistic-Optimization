```python
df1 = pd.read_csv('历史轨迹7.21.csv')
df2 = pd.read_csv('Xiamen_Unique_locations最新有仓库.csv')
```

### 200m = 0.001796 degree in longitude or lattitude
### 100m = 0.000898 degree
### 50m = 0.000448 degree
### 25m = 0.000225 degree
### 22m = 0.0002 degree

```python
# Assuming df1 and df2 are already loaded
# Decompose the 'location' column in df1 and df2 into separate 'longitude' and 'latitude' columns
df1['longitude'] = df1['location'].str.split(',').str[0].astype(float)
df1['latitude'] = df1['location'].str.split(',').str[1].astype(float)

df2['longitude'] = df2['location'].str.split(',').str[0].astype(float)
df2['latitude'] = df2['location'].str.split(',').str[1].astype(float)

# Define the range
delta = 0.000898

# Create the bounds for df2
df2['longitude_min'] = df2['longitude'] - delta
df2['longitude_max'] = df2['longitude'] + delta
df2['latitude_min'] = df2['latitude'] - delta
df2['latitude_max'] = df2['latitude'] + delta

# Check if each df1 location is within the square bounds of any df2 location
def check_within_bounds(row):
    matching_rows = df2[
        (df2['longitude_min'] <= row['longitude']) & 
        (df2['longitude_max'] >= row['longitude']) & 
        (df2['latitude_min'] <= row['latitude']) & 
        (df2['latitude_max'] >= row['latitude'])
    ]
    if not matching_rows.empty:
        return matching_rows.iloc[0][['TRANSID', 'INCEPTADDR']]
    else:
        return pd.Series([None, None], index=['TRANSID', 'INCEPTADDR'])

results = df1.apply(check_within_bounds, axis=1)
df1['TRANSID'] = results['TRANSID']
df1['INCEPTADDR'] = results['INCEPTADDR']

# Drop the extra columns we added
df1 = df1.drop(['longitude', 'latitude'], axis=1)
# Drop NA Inceptaddr
filtered_df = df1[df1['INCEPTADDR'].notna()]
filtered_df
```

```python
# Save filtered_df to csv file
filtered_df.to_csv('历史轨迹9.1-all_200m.csv', index=False, encoding='utf-8-sig')
```

```python
# Work on a copy to avoid SettingWithCopyWarning
filtered_df = pd.read_csv("历史轨迹9.1-停车点100m.csv")

# Convert "定位时间" column to datetime
filtered_df['定位时间'] = pd.to_datetime(filtered_df['定位时间'])

# Extract numerical value from "行驶里程" and convert it to a float
filtered_df['行驶里程'] = filtered_df['行驶里程'].str.extract('(\d+\.\d+)').astype(float)

# Compute the difference for "行驶里程" and "定位时间"
filtered_df['distance'] = filtered_df['行驶里程'].diff().fillna(0) * 1000  # Convert from km to m for the difference
filtered_df['duration'] = filtered_df['定位时间'].diff().fillna(pd.Timedelta(seconds=0)).dt.total_seconds()

# For the first row, set "distance" and "duration" to 0
filtered_df.iloc[0, filtered_df.columns.get_loc('distance')] = 0
filtered_df.iloc[0, filtered_df.columns.get_loc('duration')] = 0

# Unique for TRANSID
filtered_df = filtered_df.drop_duplicates(subset='TRANSID', keep='first').reset_index(drop=True)

filtered_df
```

```python
# Shift the columns to pair each row with its next row
filtered_df['destination'] = filtered_df['TRANSID'].shift(-1)
filtered_df['destination_location'] = filtered_df['location'].shift(-1)

# Create the new dataframe df_91
df_91 = filtered_df[['TRANSID', 'location', 'destination', 'destination_location']].copy()
df_91.columns = ['origin', 'origin_location', 'destination', 'destination_location']

# Remove the last row since it doesn't have a destination (because of the shift)
df_91 = df_91[:-1].reset_index(drop=True)
shifted_df = filtered_df.shift(-1)

# Create df_91 using the data from filtered_df and shifted_df
df_91 = pd.DataFrame({
    'origin': filtered_df['TRANSID'],
    'destination': shifted_df['TRANSID'],
    'origin_location': filtered_df['location'],
    'destination_location': shifted_df['location'],
    'distance_r': shifted_df['distance'],
    'duration_r': shifted_df['duration']
})

# Drop the last row since it will have NaN values (because of the shift operation)
df_91 = df_91.dropna().reset_index(drop=True)

df_91
```

```python
# Save filtered_df to csv file
df_91.to_csv('9.1-停车点100mlocations.csv', index=False, encoding='utf-8-sig')
```

### Compare real distance and duration with the api results

```python
df_cmp = pd.read_csv("9.1-停车点200mlocations_nd.csv")
# Plotting distance vs distance_r
plt.figure(figsize=(14,6))

plt.subplot(1, 2, 1)
plt.plot(df_cmp.index, df_cmp['distance'], label='distance', color='blue')
plt.plot(df_cmp.index, df_cmp['distance_r'], label='distance_r', color='red')
plt.xlabel('Index')
plt.ylabel('Distance Value')
plt.legend()
plt.title('Comparison between distance and distance_r')

# Plotting duration vs duration_r
plt.subplot(1, 2, 2)
plt.plot(df_cmp.index, df_cmp['duration'], label='duration', color='blue')
plt.plot(df_cmp.index, df_cmp['duration_r'], label='duration_r', color='red')
plt.xlabel('Index')
plt.ylabel('Duration Value')
plt.legend()
plt.title('Comparison between duration and duration_r')

plt.tight_layout()
plt.show()
```

#### Several points shows high difference on distance and duration. This may causes by 两个地点不是前后的送货点，中间有些地点没有被找到；时间差距太大可能司机中途有做别的时间，并不是直接接着送下一个地点的

```python
df_cmp = df_cmp[(df_cmp['distance_r'] <= 40000) & (df_cmp['duration_r'] <= 40000)]

# Plotting distance vs distance_r
plt.figure(figsize=(14,6))

plt.subplot(1, 2, 1)
plt.plot(df_cmp.index, df_cmp['distance'], label='distance', color='blue')
plt.plot(df_cmp.index, df_cmp['distance_r'], label='distance_r', color='red')
plt.xlabel('Index')
plt.ylabel('Distance Value')
plt.legend()
plt.title('Comparison between distance and distance_r')

# Plotting duration vs duration_r
plt.subplot(1, 2, 2)
plt.plot(df_cmp.index, df_cmp['duration'], label='duration', color='blue')
plt.plot(df_cmp.index, df_cmp['duration_r'], label='duration_r', color='red')
plt.xlabel('Index')
plt.ylabel('Duration Value')
plt.legend()
plt.title('Comparison between duration and duration_r')

plt.tight_layout()
plt.show()
```

```python
# Ratio between real distance and gaode_api
df_cmp['distance_ratio'] = df_cmp['distance_r'] / df_cmp['distance']
average_distance_ratio = df_cmp['distance_ratio'].mean()
average_distance_ratio
```
