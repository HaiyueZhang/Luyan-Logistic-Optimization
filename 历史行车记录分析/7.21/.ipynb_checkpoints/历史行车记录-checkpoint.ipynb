```python
# First, ensure that the '定位时间' column is in datetime format
df1['定位时间'] = pd.to_datetime(df1['定位时间'], format='%Y-%m-%d %H:%M:%S')

# Now, group by '车牌号码'
grouped = df1.groupby('车牌号码')

# Define a function to apply to each group to get the required values
def get_stats(group):
    return pd.Series({
        'Total Distance': group['行驶里程'].max(),
        'Start Time': group['定位时间'].min(),
        'End Time': group['定位时间'].max(),
        'Time Duration': group['定位时间'].max() - group['定位时间'].min()
    })

# Apply the function to each group
stats = grouped.apply(get_stats)

# Reset index to have '车牌号码' as a column
stats = stats.reset_index()

# Now, you can print or examine 'stats' DataFrame
print(stats)

# Remove the 'km' from the 'Total Distance' and convert the column to float
stats['Total Distance'] = stats['Total Distance'].replace(' km', '', regex=True).astype(float)
# Sum up the 'Total Distance' column
total_distance_sum = stats['Total Distance'].sum()
print(f"The sum of the Total Distance is: {total_distance_sum} km")
```

```python
# Remove rows where "已停留/总停留" is empty or just has "/"
df1_non_empty = df1[df1['已停留/总停留'].astype(str).str.strip().str.match(r'.+/.+')]

# Find the difference between consecutive "序号" values
df1_non_empty['diff'] = df1_non_empty['序号'].diff().fillna(1)

# A sequence starts whenever the difference is not equal to 1
df1_non_empty['sequence'] = df1_non_empty['diff'].ne(1).cumsum()

# Group by 'sequence' and compute the sequence count
df1_non_empty['sequence_count'] = df1_non_empty.groupby('sequence').cumcount() + 1

# Filter sequences where the maximum 'sequence_count' is at least 4
sequences_with_min_length = df1_non_empty.groupby('sequence')['sequence_count'].transform('max') >= 4

# Apply this filter to the DataFrame
df1_filtered = df1_non_empty[sequences_with_min_length]

df1_filtered
```

```python
df1_filtered = pd.read_csv('7.21连续停靠点drop.csv')
# Function to convert the time format "X时Y分Z秒" into seconds
def convert_to_seconds(time_str):
    if pd.notna(time_str):  # Check if the time_str is not NaN
        # Initialize hours, minutes, and seconds to 0
        hours, minutes, seconds = 0, 0, 0
        
        # Check if '时' is present
        if '时' in time_str:
            parts = time_str.split('时')
            hours = int(parts[0])
            time_str = parts[1]  # Remaining part after '时'
        
        # Check if '分' is present
        if '分' in time_str:
            parts = time_str.split('分')
            minutes = int(parts[0])
            # Check if there's a second part after '分' (for seconds)
            if len(parts) > 1 and parts[1] != '':
                seconds = int(parts[1].rstrip('秒'))
        else:
            # If '分' is not present, then we only have seconds
            seconds = int(time_str.rstrip('秒'))
            
        return hours * 3600 + minutes * 60 + seconds
    else:
        return 0  # Return 0 if the input is NaN

# Extract the total duration part (after the slash) for each group's first row
# Split by ' / ' (space, slash, space) and get the second part [1]
df1_filtered['total_duration'] = df1_filtered.groupby('sequence')['已停留/总停留'].transform('first').str.split(' / ').str[1]

# Convert the total duration to seconds using the conversion function
df1_filtered['total_duration_seconds'] = df1_filtered['total_duration'].apply(convert_to_seconds)

# Group by 'sequence' and calculate the average duration in seconds for each group
df_group_avg_duration = df1_filtered.groupby('sequence')['total_duration_seconds'].mean().reset_index()

# Rename the columns for clarity
df_group_avg_duration.columns = ['sequence', 'average_duration_seconds']
df_group_avg_duration
```

```python
# Filter the dataframe where 'average_duration_seconds' is less than 2400
filtered_df = df_group_avg_duration[df_group_avg_duration['average_duration_seconds'] < 3600]

# Calculate the sum of 'average_duration_seconds' for the filtered rows
sum_of_durations = filtered_df['average_duration_seconds'].sum()

# Calculate the count of the filtered rows
count_of_durations = len(filtered_df)

# Calculate the average duration for the filtered rows
# Here we check if count_of_durations is not zero to avoid division by zero
average_duration = sum_of_durations / count_of_durations if count_of_durations else 0

print(f"Average Duration for filtered groups: {average_duration} seconds")
filtered_df
```

```python
# Assuming df1 and df2 are already loaded
# Decompose the 'location' column in df1 and df2 into separate 'longitude' and 'latitude' columns
df1['longitude'] = df1['location'].str.split(',').str[0].astype(float)
df1['latitude'] = df1['location'].str.split(',').str[1].astype(float)

df2['longitude'] = df2['location'].str.split(',').str[0].astype(float)
df2['latitude'] = df2['location'].str.split(',').str[1].astype(float)

# Define the range
delta = 0.000898

# Create the bounds for df2
df2['longitude_min'] = df2['longitude'] - delta
df2['longitude_max'] = df2['longitude'] + delta
df2['latitude_min'] = df2['latitude'] - delta
df2['latitude_max'] = df2['latitude'] + delta

# Check if each df1 location is within the square bounds of any df2 location
def check_within_bounds(row):
    matching_rows = df2[
        (df2['longitude_min'] <= row['longitude']) & 
        (df2['longitude_max'] >= row['longitude']) & 
        (df2['latitude_min'] <= row['latitude']) & 
        (df2['latitude_max'] >= row['latitude'])
    ]
    if not matching_rows.empty:
        return matching_rows.iloc[0][['TRANSID', 'INCEPTADDR']]
    else:
        return pd.Series([None, None], index=['TRANSID', 'INCEPTADDR'])

results = df1.apply(check_within_bounds, axis=1)
df1['TRANSID'] = results['TRANSID']
df1['INCEPTADDR'] = results['INCEPTADDR']

# Drop the extra columns we added
df1 = df1.drop(['longitude', 'latitude'], axis=1)
# Drop NA Inceptaddr
filtered_df = df1[df1['INCEPTADDR'].notna()]
filtered_df
```

```python
# Work on a copy to avoid SettingWithCopyWarning
filtered_df = pd.read_csv("历史轨迹9.1-停车点100m.csv")

# Convert "定位时间" column to datetime
filtered_df['定位时间'] = pd.to_datetime(filtered_df['定位时间'])

# Extract numerical value from "行驶里程" and convert it to a float
filtered_df['行驶里程'] = filtered_df['行驶里程'].str.extract('(\d+\.\d+)').astype(float)

# Compute the difference for "行驶里程" and "定位时间"
filtered_df['distance'] = filtered_df['行驶里程'].diff().fillna(0) * 1000  # Convert from km to m for the difference
filtered_df['duration'] = filtered_df['定位时间'].diff().fillna(pd.Timedelta(seconds=0)).dt.total_seconds()

# For the first row, set "distance" and "duration" to 0
filtered_df.iloc[0, filtered_df.columns.get_loc('distance')] = 0
filtered_df.iloc[0, filtered_df.columns.get_loc('duration')] = 0

# Unique for TRANSID
filtered_df = filtered_df.drop_duplicates(subset='TRANSID', keep='first').reset_index(drop=True)

filtered_df
```

```python
# Shift the columns to pair each row with its next row
filtered_df['destination'] = filtered_df['TRANSID'].shift(-1)
filtered_df['destination_location'] = filtered_df['location'].shift(-1)

# Create the new dataframe df_91
df_91 = filtered_df[['TRANSID', 'location', 'destination', 'destination_location']].copy()
df_91.columns = ['origin', 'origin_location', 'destination', 'destination_location']

# Remove the last row since it doesn't have a destination (because of the shift)
df_91 = df_91[:-1].reset_index(drop=True)
shifted_df = filtered_df.shift(-1)

# Create df_91 using the data from filtered_df and shifted_df
df_91 = pd.DataFrame({
    'origin': filtered_df['TRANSID'],
    'destination': shifted_df['TRANSID'],
    'origin_location': filtered_df['location'],
    'destination_location': shifted_df['location'],
    'distance_r': shifted_df['distance'],
    'duration_r': shifted_df['duration']
})

# Drop the last row since it will have NaN values (because of the shift operation)
df_91 = df_91.dropna().reset_index(drop=True)

df_91
```

```python
df_cmp = pd.read_csv("9.1-停车点200mlocations_nd.csv")
# Plotting distance vs distance_r
plt.figure(figsize=(14,6))

plt.subplot(1, 2, 1)
plt.plot(df_cmp.index, df_cmp['distance'], label='distance', color='blue')
plt.plot(df_cmp.index, df_cmp['distance_r'], label='distance_r', color='red')
plt.xlabel('Index')
plt.ylabel('Distance Value')
plt.legend()
plt.title('Comparison between distance and distance_r')

# Plotting duration vs duration_r
plt.subplot(1, 2, 2)
plt.plot(df_cmp.index, df_cmp['duration'], label='duration', color='blue')
plt.plot(df_cmp.index, df_cmp['duration_r'], label='duration_r', color='red')
plt.xlabel('Index')
plt.ylabel('Duration Value')
plt.legend()
plt.title('Comparison between duration and duration_r')

plt.tight_layout()
plt.show()
```
