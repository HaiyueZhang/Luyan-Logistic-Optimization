```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from datetime import datetime
import numpy as np
```

```python
file_path = "全部车辆-7月20-21-22-轨迹数据.xlsx"
df1 = pd.read_excel(file_path, engine='openpyxl')
df1
```

```python
df1 = df1.drop(columns=["Unnamed: 0"])

# Drop the first row
df1 = df1.iloc[1:]

# Set the second row as column names
df1.columns = df1.iloc[0]
df1 = df1.iloc[1:]

# Reset the index
df1.reset_index(drop=True, inplace=True)
# Assuming df1 is your dataframe

# List of columns to be dropped
columns_to_drop = ["定位信号", "通讯信号", "终端标识", "SIM卡号", "ACC ", "定位类型"]

# Drop the columns
df1 = df1.drop(columns=columns_to_drop)
df1['location'] = df1['定位经度'].astype(str) + "," + df1['定位纬度'].astype(str)
df1 = df1.drop(columns = ['定位经度', '定位纬度'])
df1
```

```python
df1.to_csv('历史轨迹7.20-22.csv', index=False, encoding='utf-8-sig')
```

```python
df1 = pd.read_csv('历史轨迹7.20-22.csv')
excluded_values = ['闽DL9N83', '闽DT5L33']
df1 = df1[~df1['车牌号码'].isin(excluded_values)]
# df2 = pd.read_csv('Xiamen_Unique_locations最新有仓库.csv')
df3 = pd.read_csv('7_21_address.csv')
```

```python
# Split 'location' into 'longitude' and 'latitude' for df1
df1[['longitude', 'latitude']] = df1['location'].str.split(',', expand=True).astype(float)

# Do the same for df3
df3[['longitude', 'latitude']] = df3['combine'].str.split(',', expand=True).astype(float)

# Plotting with a larger figure size, e.g., 15 inches by 10 inches
fig, ax = plt.subplots(figsize=(13, 10))

# Generate a color for each '车牌号码' group in df1
colors = plt.cm.jet(np.linspace(0, 1, df1['车牌号码'].nunique()))

# Group by '车牌号码' and plot for df1
for (license_plate), group_data in df1.groupby('车牌号码'):
    ax.scatter(group_data['longitude'], group_data['latitude'], label=license_plate, s=10)

# Plot data from df3, using a different color and/or marker
# You may choose a specific color or marker for df3 to differentiate
ax.scatter(df3['longitude'], df3['latitude'], color='red', marker='^', label='df3', s=10)

# Define the range for the rectangle
long_min, long_max = 117.882223, 118.373857
lat_min, lat_max = 24.41, 24.905

# Create a Rectangle patch
rect = patches.Rectangle((long_min, lat_min), long_max - long_min, lat_max - lat_min,
                         linewidth=1, edgecolor='r', facecolor='none')

# Add the rectangle to the plot
ax.add_patch(rect)

# Optionally set a grid
ax.grid(True)

# Optionally add a legend outside of the plot
ax.legend(title='车牌号码', bbox_to_anchor=(1.05, 1), loc='upper left')

# Set labels for the axes
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')

# Set title for the plot
ax.set_title('Locations by 车牌号码 with Highlighted Area')

# Show the plot
plt.show()
```
