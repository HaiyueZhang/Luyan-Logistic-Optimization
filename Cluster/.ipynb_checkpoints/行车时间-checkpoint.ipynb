```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster
```

```python
# Read the CSV file
df_tms = pd.read_csv("y.csv", encoding="utf-8")  # assuming the file is encoded in UTF-8
df_Xiamen = pd.read_csv("new_onlyXiamen_addcode.csv", encoding="utf-8")
```

```python
# Filter columns
df_tms = df_tms[['ROUTEID', 'TRANSID','SOURCEID','SENDPOS', 'DISPATCHDATE', 'SIGNDATE']]
# Drop NA
df_tms = df_tms.dropna(subset=['DISPATCHDATE', 'SIGNDATE'])
# Acquire duration time
df_tms['SIGNDATE'] = pd.to_datetime(df_tms['SIGNDATE'], format='%Y/%m/%d %H:%M:%S')
df_tms['DISPATCHDATE'] = pd.to_datetime(df_tms['DISPATCHDATE'], format='%Y/%m/%d %H:%M:%S')
df_tms['DURATION'] = df_tms['SIGNDATE'] - df_tms['DISPATCHDATE']
df_tms['DURATION'] = df_tms['DURATION'].dt.total_seconds() / 60  # Convert timedelta to minutes
# Filter out rows where 'DURATION' is less than 1
# df_tms = df_tms[df_tms['DURATION'] >= 1]
# Extract the date from the 'SIGNDATE' column and create a new 'DATE' column
df_tms['DATE'] = df_tms['SIGNDATE'].dt.date
# Merge df_Xiamen with df_tms to filter all the trans in Xiamen
merged_df = df_tms.merge(df_Xiamen[['TransID','Xiamen.经度', 'Xiamen.纬度', 'combine']], left_on='TRANSID', right_on='TransID', how='inner')
merged_df.drop(columns=['TransID'], inplace=True)
merged_df
```

```python
# Analysis specific date in XiaMen
df_date = merged_df
df_date_count = df_date.groupby('DATE').size().reset_index(name='Count')

# Convert 'DATE' column to datetime for comparison
df_date_count['DATE'] = pd.to_datetime(df_date_count['DATE'])

# Plot a line chart between DATE and Count
plt.figure(figsize=(10, 6))
plt.plot(df_date_count['DATE'], df_date_count['Count'], marker='o', label='All Dates')

# Highlight specific dates with red dots
specific_dates = pd.to_datetime(['2023-05-02', '2023-05-03', '2023-05-07', '2023-05-13', '2023-05-14', '2023-05-20', '2023-05-21', '2023-05-27', '2023-05-28',
                                 '2023-06-03', '2023-06-04', '2023-06-10', '2023-06-11', '2023-06-17', '2023-06-18', '2023-06-22', '2023-06-23', '2023-06-24', 
                                 '2023-07-01', '2023-07-02', '2023-07-08', '2023-07-09', '2023-07-15', '2023-07-16', '2023-07-22', '2023-07-23', '2023-07-29', '2023-07-30', 
                                 '2023-08-05', '2023-08-06', '2023-08-12', '2023-08-13', '2023-08-19', '2023-08-20', '2023-08-26', '2023-08-27'])
for date in specific_dates:
    mask = df_date_count['DATE'] == date
    if mask.sum() > 0:
        count = df_date_count.loc[mask, 'Count'].values[0]
        plt.plot(date, count, marker='o', color='red', linestyle='None', label='Weekend and Holiday')

# Adding title, labels, grid, and legend
plt.title('Count by Date')
plt.xlabel('Date')
plt.ylabel('Count')
plt.grid(True)
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust spacing for labels

# Ensure unique legend entries
handles, labels = plt.gca().get_legend_handles_labels()
by_label = dict(zip(labels, handles))
plt.legend(by_label.values(), by_label.keys())

plt.show()
```

```python
df_date = merged_df
df_date_count = df_date.groupby('DATE').size().reset_index(name='Count')

# Convert 'DATE' column to datetime for comparison
df_date_count['DATE'] = pd.to_datetime(df_date_count['DATE'])

# Remove specific dates from df_date_count
specific_dates = pd.to_datetime(['2023-05-02', '2023-05-03', '2023-05-07', '2023-05-13', '2023-05-14', '2023-05-20', '2023-05-21', '2023-05-27', '2023-05-28',
                                 '2023-06-03', '2023-06-04', '2023-06-10', '2023-06-11', '2023-06-17', '2023-06-18', '2023-06-22', '2023-06-23', '2023-06-24', 
                                 '2023-07-01', '2023-07-02', '2023-07-08', '2023-07-09', '2023-07-15', '2023-07-16', '2023-07-22', '2023-07-23', '2023-07-29', '2023-07-30', 
                                 '2023-08-05', '2023-08-06', '2023-08-12', '2023-08-13', '2023-08-19', '2023-08-20', '2023-08-26', '2023-08-27'])
df_date_count = df_date_count[~df_date_count['DATE'].isin(specific_dates)]

# Filter rows in DataFrame where 'DATE' is up to '2023-08-15'
end_date = pd.to_datetime('2023-08-15')
df_date_count = df_date_count[df_date_count['DATE'] <= end_date]

# Calculate the average count, variance, and standard deviation
average_count = df_date_count['Count'].mean()
variance_count = df_date_count['Count'].var()
std_dev_count = df_date_count['Count'].std()

# Plot a line chart between DATE and Count for the remaining dates
plt.figure(figsize=(10, 6))
plt.plot(df_date_count['DATE'], df_date_count['Count'], marker='o', label='All Dates')

# Plot the average line
plt.axhline(y=average_count, color='r', linestyle='--', label=f'Average Count: {average_count:.2f}')

# Adding title, labels, grid, and legend
plt.title('Count by Date (Excluding Specific Dates)')
plt.xlabel('Date')
plt.ylabel('Count')
plt.grid(True)
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust spacing for labels
plt.legend()  # Add legend to the plot

# Annotate variance and standard deviation on the plot
plt.annotate(f'Variance: {variance_count:.2f}', xy=(0.05, 0.95), xycoords='axes fraction')
plt.annotate(f'Standard Deviation: {std_dev_count:.2f}', xy=(0.05, 0.90), xycoords='axes fraction')

plt.show()
```

### The standard deviation is 141.51, which is quite high and means that there is more variability in the number of deliveries each day.

```python
# Groupby Date and RouteID
df_dr = merged_df
df_dr = df_dr.groupby(['DATE', 'ROUTEID']).size().reset_index(name='Count')
# Get unique RouteIDs
unique_route_ids = sorted(df_dr['ROUTEID'].unique())

# Set up the figure and axes
fig, axs = plt.subplots(len(unique_route_ids), figsize=(10, 6*len(unique_route_ids)))

# Loop over each unique and sorted RouteID and plot a line chart
for i, route_id in enumerate(unique_route_ids):
    df_route = df_dr[df_dr['ROUTEID'] == route_id]
    
    axs[i].plot(df_route['DATE'], df_route['Count'], marker='o', label=f'Route ID: {route_id}')
    axs[i].set_title(f'Count by Date for Route ID: {route_id}')
    axs[i].set_xlabel('Date')
    axs[i].set_ylabel('Count')
    axs[i].grid(True)
    axs[i].legend()

# Adjust layout
plt.tight_layout()
plt.show()
```
