```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read the CSV file
df_tms = pd.read_csv("TMS派车装载管理820001.csv", encoding="utf-8", low_memory=False)
df_Xiamen = pd.read_csv("new_onlyXiamen_addcode.csv", encoding="utf-8")
df_history = pd.read_excel("历史轨迹.xlsx", header=1)
```

```python
# Filter columns
df_tms = df_tms[['ROUTEID', 'TRANSID', 'SENDPOS', 'DISPATCHDATE', 'SIGNDATE']]
# Drop NA
df_tms = df_tms.dropna(subset=['DISPATCHDATE', 'SIGNDATE'])
# Extract the date from the 'SIGNDATE' column and create a new 'DATE' column
df_tms['SIGNDATE'] = pd.to_datetime(df_tms['SIGNDATE'], format='%m/%d/%Y %H:%M')
df_tms['DATE'] = df_tms['SIGNDATE'].dt.date
merged_df = df_tms.merge(df_Xiamen[['TransID','Xiamen.经度', 'Xiamen.纬度', 'combine']], left_on='TRANSID', right_on='TransID', how='inner')
merged_df.drop(columns=['TransID'], inplace=True)
merged_df
```

```python
# Analysis specific date in XiaMen
df_date = merged_df
df_date_count = df_date.groupby('DATE').size().reset_index(name='Count')

# Convert 'DATE' column to datetime for comparison
df_date_count['DATE'] = pd.to_datetime(df_date_count['DATE'])

# Plot a line chart between DATE and Count
plt.figure(figsize=(10, 6))
plt.plot(df_date_count['DATE'], df_date_count['Count'], marker='o', label='All Dates')

# Highlight specific dates with red dots
specific_dates = pd.to_datetime(['2023-05-02', '2023-05-03', '2023-05-07', '2023-05-13', '2023-05-14', '2023-05-20', '2023-05-21', '2023-05-27', '2023-05-28',
                                 '2023-06-03', '2023-06-04', '2023-06-10', '2023-06-11', '2023-06-17', '2023-06-18', '2023-06-22', '2023-06-23', '2023-06-24', 
                                 '2023-07-01', '2023-07-02', '2023-07-08', '2023-07-09', '2023-07-15', '2023-07-16', '2023-07-22', '2023-07-23', '2023-07-29', '2023-07-30', 
                                 '2023-08-05', '2023-08-06', '2023-08-12', '2023-08-13', '2023-08-19', '2023-08-20', '2023-08-26', '2023-08-27',
                                 '2023-09-02', '2023-09-03', '2023-09-09', '2023-09-10', '2023-09-16', '2023-09-17'])
for date in specific_dates:
    mask = df_date_count['DATE'] == date
    if mask.sum() > 0:
        count = df_date_count.loc[mask, 'Count'].values[0]
        plt.plot(date, count, marker='o', color='red', linestyle='None', label='Weekend and Holiday')

# Adding title, labels, grid, and legend
plt.title('Count by Date')
plt.xlabel('Date')
plt.ylabel('Count')
plt.grid(True)
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust spacing for labels

# Ensure unique legend entries
handles, labels = plt.gca().get_legend_handles_labels()
by_label = dict(zip(labels, handles))
plt.legend(by_label.values(), by_label.keys())

plt.show()
```

```python
df_newtms = pd.read_csv("TMS派车装载管理820001.csv", encoding="utf-8", low_memory=False)
# Filter columns
df_newtms = df_newtms[['ROUTEID', 'TRANSID', 'SENDPOS', 'CREDATE', 'DISPATCHDATE', 'SIGNDATE']]
# Drop NA
df_newtms = df_newtms.dropna(subset=['CREDATE'])
# Extract the date from the 'SIGNDATE' column and create a new 'DATE' column
df_newtms['CREDATE'] = pd.to_datetime(df_newtms['CREDATE'], format='%m/%d/%Y %H:%M')
df_newtms['DATE'] = df_newtms['CREDATE'].dt.date
df_newtms['DATE'] = pd.to_datetime(df_newtms['DATE'])
# Filter Xiamen address
merged_df_new = df_newtms.merge(df_Xiamen[['TransID','Xiamen.经度', 'Xiamen.纬度', 'combine']], left_on='TRANSID', right_on='TransID', how='inner')
merged_df_new.drop(columns=['TransID'], inplace=True)

# Plot
# Convert the DATE column to datetime format
merged_df_new['DATE'] = pd.to_datetime(merged_df_new['DATE'])

# Group by the DATE column and count the number of rows for each date
date_counts = merged_df_new.groupby('DATE').size()

# Specify the dates you want to highlight in red
highlight_dates = pd.to_datetime(['2023-07-09', '2023-07-16', '2023-07-22', '2023-07-23', '2023-07-29', '2023-07-30', 
                                 '2023-08-05', '2023-08-06', '2023-08-13', '2023-08-20', '2023-08-26', 
                                 '2023-09-02', '2023-09-03', '2023-09-09', '2023-09-10', '2023-09-17'])  # Replace with your specific dates

# Find the counts for the highlight dates
highlight_counts = date_counts[highlight_dates]

# Plot the line chart with dots for each date
plt.figure(figsize=(10,6))
date_counts.plot(kind='line', marker='o')

# Plot the red dots for the specific highlight dates
plt.scatter(highlight_dates, highlight_counts, color='red', s=100, label='Highlighted Dates')

plt.title('Number of Rows for Each Date')
plt.xlabel('Date')
plt.ylabel('Count')
plt.legend()
plt.show()

```

```python
# Convert the DATE column to datetime format
merged_df_new['DATE'] = pd.to_datetime(merged_df_new['DATE'])

# Group by the DATE column and count the number of rows for each date
date_counts = merged_df_new.groupby('DATE').size()

# Specify the dates you want to exclude
exclude_dates = pd.to_datetime(['2023-07-09', '2023-07-16', '2023-07-22', '2023-07-23', '2023-07-29', '2023-07-30', 
                                '2023-08-05', '2023-08-06', '2023-08-13', '2023-08-20', '2023-08-26', 
                                '2023-09-02', '2023-09-03', '2023-09-09', '2023-09-10', '2023-09-17'])

# Filter date_counts to exclude the specified dates
filtered_date_counts = date_counts.drop(exclude_dates, errors='ignore')

# Calculate the mean, standard deviation, and variance of the filtered date counts
mean_count = filtered_date_counts.mean()
std_dev_count = filtered_date_counts.std()
variance_count = filtered_date_counts.var()

# Print the calculated values
print("Mean:", mean_count)
print("Standard Deviation:", std_dev_count)
print("Variance:", variance_count)

# Plot the line chart with dots for each date in the filtered data
plt.figure(figsize=(10,6))
filtered_date_counts.plot(kind='line', marker='o')

# Add a horizontal line representing the mean
plt.axhline(y=mean_count, color='r', linestyle='--', label=f'Average: {mean_count:.2f}')

plt.title('Number of Rows for Each Date (Excluding Specific Dates)')
plt.xlabel('Date')
plt.ylabel('Count')
plt.legend()
plt.show()
```
